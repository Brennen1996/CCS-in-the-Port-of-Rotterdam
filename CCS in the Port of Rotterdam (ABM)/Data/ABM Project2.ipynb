```python
#Load the pandas library needed for dataframes
import pandas as pd
```

```python
df_percentage_port = pd.read_csv("final project v11 Percentage port-table.csv", delimiter = ",", header = 6)
```

```python
df_percentage_port.head()
```

```python
plot = df_percentage_port.groupby(["percentage-port"])["total-co2"].mean().plot(title = "Total CO2 and percentage port")

plot.set_ylabel("Total CO2 (Mt)")
plot.set_xlabel("Percentage of subsidy to the port (%)")
```

```python
plot = df_percentage_port.groupby(["percentage-port"])["total-stored-co2"].mean().plot(title = "Total Stored CO2 and percentage port")

plot.set_ylabel("Total Stored CO2 (Mt)")
plot.set_xlabel("Percentage of subsidy to the port (%)")
```

```python
plot = df_percentage_port.groupby(["percentage-port"])["total-emitted-co2"].mean().plot(title = "Total Emitted CO2 and percentage port")

plot.set_ylabel("Total Emitted CO2 (Mt)")
plot.set_xlabel("Percentage of subsidy to the port (%)")
```

```python
plot = df_percentage_port.groupby(["percentage-port"])["total-subsidy-port"].mean().plot(title = "Total subsidy port and percentage port")

plot.set_ylabel("Total Subsidy Port (M EUR)")
plot.set_xlabel("Percentage of subsidy to the port (%)")
```

```python
plot = df_percentage_port.groupby(["percentage-port"])["total-subsidy-industries"].mean().plot(title = "Total subsidy industries and percentage port")

plot.set_ylabel("Total Subsidy Industries (M EUR)")
plot.set_xlabel("Percentage of subsidy to the port (%)")
```

```python
plot = df_percentage_port.groupby(["percentage-port"])["total-electricity-use"].mean().plot(title = "Total electricity use and percentage port")

plot.set_ylabel("Total Electricity Use (MWh)")
plot.set_xlabel("Percentage of subsidy to the port (%)")
```
